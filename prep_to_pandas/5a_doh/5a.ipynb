```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
wb=openpyxl.load_workbook('O:\PP-PROLOG\VENDOR_MANAGEMENT\Data_Analytics\Data_Files\Daily_Reports\PL_INV_BY_BASE.xlsx')
```

```python
#wb.active is the first worksheet
ws=wb.active
data=ws.values
columns=next(data)[0:]

# raw_df contains all original data from PL_INV_BY_BASE
raw_df=pd.DataFrame(data,columns=columns)
convert_dict = {'IOH - Item Descr' : str,
                'IOH - Qty On Hand': float,
                'IOH - On Hand Value': float,
                'IOH - LTM Demand Qty': float,
                'IOH - LTM Demand Value': float,
                'IOH - Reorder Point': float,
                'IOH - Reorder Qty': float,
                'IOH - Max Qty': float
                }
 
raw_df = raw_df.astype(convert_dict)
raw_df.head()
```

```python
# do computations on a copy to avoid chained assignment
df_cs04reorder=raw_df[raw_df['IOH - Unit']=='CS004'].copy()
df_cs04reorder.head()
```

```python
# Aggregate on several fields
df_cs04reorder = df_cs04reorder.groupby(['IOH - Item ID', 'IOH - Utilization Type', 'IOH - Current Status', 'IOH - No Repl Flag']).agg(
 {'IOH - Max Qty': 'sum' , 'IOH - Reorder Point':'sum', 'IOH - Reorder Qty':'sum', 'IOH - LTM Demand Qty':'sum', 'IOH - LTM Demand Value':'sum'}       
).reset_index()

rename_dict = {'IOH - Reorder Qty':'CS04 Reorder Qty', 'IOH - Item ID':'item_id', 'IOH - Current Status':'CS04 Current Status', 'IOH - Reorder Point':'CS04 Reorder Point', 
                'IOH - Max Qty': 'CS04 Max Qty', 'IOH - No Repl Flag':'CS04 No Repl Flag', 'IOH - Utilization Type':'CS04 Utilization Type', 'IOH - LTM Demand Qty':'CS04 LTM Demand Qty', 
                'IOH - LTM Demand Value': 'CS04 LTM Demand Value'}

# call rename () method
df_cs04reorder.rename(columns= rename_dict, inplace=True)
```

```python
df_cs04reorder['CS04 Do Not Replenish']=df_cs04reorder.apply(lambda x: (x['CS04 Reorder Point']==0 and x['CS04 Reorder Qty']==0 and ['CS04 Max Qty']==0) 
                                                                        or (str(x['CS04 No Repl Flag'])=='Y') or (str(x['CS04 Current Status'])=='Discnt'), axis = 1)
                                                                        
df_cs04reorder['No CS004 Min Set']=df_cs04reorder.apply(lambda x: (x['CS04 Reorder Point']==0), axis = 1)

```

```python
# do computations on a copy to avoid chained assignment
df_latestioh=raw_df[raw_df['IOH - Unit']!='CS003'].copy()
df_latestioh.head()
```

```python
#All the calculations for the clean 1 step
def location_type_apply(x):
    if x[0:1] == "C":
        return "CS"
    elif int(x[-2:]) <= 20:
        return "Bus" 
    elif int(x[-2:]) >= 21:
        return "Rail"
    else:
        return None

def part_in_type_CS004_apply(x):
    if x=='CS004':
        return 1
    else:
        return 0
    
def part_in_type_others_apply(x):
    if x=='CS001' or x=='CS002':
        return 1
    else:
        return 0

def other_cs_count_apply(x):
    if x[0:2]=='CS'and x!='CS004':
        return 1
    else:
        return 0

def base_unit_count_apply(x):
    if x[0:2] == "BS":
        return 1
    else:
        return 0
```

```python
# Create the location type based on the location_type_apply function
df_latestioh['Location Type'] = df_latestioh['IOH - Unit'].apply(location_type_apply)

#Bus
df_latestioh['Bus OH Quantity']=df_latestioh.apply(lambda x: float(x['IOH - Qty On Hand']) if x['Location Type']=='Bus' else 0, axis = 1)
df_latestioh['Bus LTM Demand Quantity']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Qty']) if x['Location Type']=='Bus' else 0, axis = 1)
df_latestioh['Bus LTM Demand Value']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Value']) if x['Location Type']=='Bus' else 0, axis = 1)

#CS
df_latestioh['CS OH Quantity']=df_latestioh.apply(lambda x: float(x['IOH - Qty On Hand']) if x['Location Type']=='CS' else 0, axis = 1)
df_latestioh['CS LTM Demand Quantity']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Qty']) if x['Location Type']=='CS' else 0, axis = 1)
df_latestioh['CS LTM Demand Value']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Value']) if x['Location Type']=='CS' else 0, axis = 1)

#Rail
df_latestioh['Rail OH Quantity']=df_latestioh.apply(lambda x: float(x['IOH - Qty On Hand']) if x['Location Type']=='Rail' else 0, axis = 1)
df_latestioh['Rail LTM Demand Quantity']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Qty']) if x['Location Type']=='Rail' else 0, axis = 1)
df_latestioh['Rail LTM Demand Value']=df_latestioh.apply(lambda x: float(x['IOH - LTM Demand Value']) if x['Location Type']=='Rail' else 0, axis = 1)

#Fuctions applies
df_latestioh['CS04 Utilization Type']=df_latestioh.apply(lambda x: x['IOH - Utilization Type'] if x['IOH - Unit']=='CS004' else 0, axis = 1)
df_latestioh['Part in CS04'] = df_latestioh['IOH - Unit'].apply(part_in_type_CS004_apply)
df_latestioh['Part in CS01/02'] = df_latestioh['IOH - Unit'].apply(part_in_type_others_apply)

#No Repl Flag Count
df_latestioh['Base No Repl Flag Count']=df_latestioh.apply(lambda x: 1 if x['IOH - No Repl Flag']=='Y' and x['IOH - Unit'][0:2]=='BS' else 0, axis = 1)
df_latestioh['CS01/02 No Repl Flag Count']=df_latestioh.apply(lambda x: 1 if x['IOH - No Repl Flag']=='Y' and x['IOH - Unit'][0:2]=='CS' else 0, axis = 1)

#other Counts
df_latestioh['Other CS Count']=df_latestioh['IOH - Unit'].apply(other_cs_count_apply)
df_latestioh['Base Unit Count'] = df_latestioh['IOH - Unit'].apply(base_unit_count_apply)

#OH value
df_latestioh['Bus OH Value']=df_latestioh.apply(lambda x:float( x['IOH - On Hand Value']) if x['Location Type']=='Bus' else 0, axis = 1)
df_latestioh['CS OH Value']=df_latestioh.apply(lambda x: float(x['IOH - On Hand Value']) if x['Location Type']=='CS' else 0, axis = 1)
df_latestioh['Rail OH Value']=df_latestioh.apply(lambda x:float( x['IOH - On Hand Value']) if x['Location Type']=='Rail' else 0, axis = 1)

#Reorder points
df_latestioh['Base Reorder Point']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][0:2]=='BS' else 0, axis = 1)
df_latestioh['Central Store Reorder Point']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][0:2]=='CS' else 0, axis = 1)

#Reorder point for Central Store and Bus(highest and lowest)
df_latestioh['Base Reorder Point (Lowest)']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][0:2]=='CS' else 0, axis = 1)
df_latestioh['Base Reorder Point (Highest)']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][0:2]=='BS' else 0, axis = 1)

df_latestioh['Central Store Reorder Point (Lowest)']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if (x['IOH - Unit'][0:2]=='BS' and int(float(x['IOH - Reorder Point'])) > 0) else 0, axis = 1)
df_latestioh['Central Store Reorder Point (Highest)']=df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][0:2]=='BS' else 0, axis = 1)

# Base Reorder Point (Highest)
df_latestioh['Base Reorder Point (Highest)'] = df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][:2] == 'BS' else 0, axis = 1)

# Replenish Counts
df_latestioh['Base Do Not Replenish Count'] = df_latestioh.apply(lambda x: 1 if (x['IOH - Unit'][:2]=='BS') and 
                                                                    ((x['IOH - Reorder Qty']==0 and  x['IOH - Reorder Point']==0 and x['IOH - Max Qty']==0)
                                                                       or (x['IOH - No Repl Flag']=='Y')
                                                                       or (x['IOH - Current Status']=='Discnt'))
                                                                   else 0, axis = 1)
df_latestioh['Other CS Do Not Replenish Count'] = df_latestioh.apply(lambda x: 1 if (x['IOH - Unit']=='CS001' or x['IOH - Unit']) and 
                                                                    ((x['IOH - Reorder Qty']==0 and  x['IOH - Reorder Point']==0 and x['IOH - Max Qty']==0)
                                                                        or (x['IOH - No Repl Flag']=='Y')
                                                                        or (x['IOH - Current Status']=='Discnt'))
                                                                    else 0, axis = 1)
df_latestioh['Other CS Do Not Replenish Flag Count'] = df_latestioh.apply(lambda x: 1 if (x['IOH - Unit']=='CS001' or x['IOH - Unit']=='CS002') and x['IOH - No Repl Flag']=='Y' else 0, axis = 1)

# Avg Base Reorder Point
df_latestioh['Avg Base Reorder Point'] = df_latestioh.apply(lambda x: float(x['IOH - Reorder Point']) if x['IOH - Unit'][:2] == 'BS' else 0, axis = 1)
```

```python
# Aggregate on several fields
df_latestioh = df_latestioh.groupby(['IOH - FMIS Family Short Descr', 'IOH - FMIS Group Short Descr', 'IOH - FMIS Item Family ID', 'IOH - Item ID', 'IOH - Item Descr', 'IOH - FMIS Item Group ID', 'IOH - FMIS Item Group', 'IOH - FMIS Item Family', 'IOH - Report Date']).agg(
 {'Avg Base Reorder Point':'mean', 'Other CS Do Not Replenish Count':'sum', 'Rail LTM Demand Value':'sum', 'CS LTM Demand Value':'sum', 'Bus LTM Demand Value':'sum',
 'Rail LTM Demand Quantity':'sum', 'CS LTM Demand Quantity':'sum', 'Bus LTM Demand Quantity':'sum', 'Part in CS04': 'sum', 'Part in CS01/02': 'sum', 'CS01/02 No Repl Flag Count':'sum',
 'Base Do Not Replenish Count':'sum', 'Other CS Do Not Replenish Count':'sum', 'Other CS Count':'sum', 'Base Unit Count':'sum', 'Central Store Reorder Point':'sum', 
 'Central Store Reorder Point (Lowest)':'min', 'Central Store Reorder Point (Highest)': 'max', 'Base Reorder Point (Lowest)':'min', 'Base Reorder Point (Highest)':'max', 
 'Base Reorder Point': 'sum', 'Base No Repl Flag Count': 'sum', 'CS OH Value': 'sum', 'Rail OH Value': 'sum', 'Bus OH Value': 'sum', 'CS OH Quantity':'sum', 'Bus OH Quantity':'sum',
 'Rail OH Quantity':'sum', 'IOH - On Hand Value': 'sum', 'IOH - Unit': 'count', 'IOH - Reorder Point': 'mean', 'IOH - Qty On Hand': 'sum', 'IOH - Current Status': 'count', 
 'IOH - No Repl Flag': 'count', 'IOH - LTM Demand Qty': 'sum', 'IOH - LTM Demand Value': 'sum' }       
).reset_index()
```

```python
def duration_apply(doh, sw_value):
    if (doh == 0):
        return 'No On Hand'
    elif (doh > 720):
        return '2+ years'
    elif (doh > 360):
        return '1-2 years'
    elif (doh > 180):
        return '6-12 months'
    elif (doh > 90):
        return '3-6 months'
    elif (doh > 30):
        return '1-3 months'
    elif (doh > 0):
        return '0-1 month'
    else:
        if (sw_value == 0):
            return 'w/No On Hand Qty'
        else:
            return 'w/On hand Qty'

def cs_category_apply(cs0102, cs04):
    if cs04 > 0:
        return 'In CS04'
    elif cs0102 == 0:
        return 'Base Only'
    elif cs0102 == 1:
        return 'In CS01 or CS02'
    else:
        return 'In both CS01/02'


rename_dict = {'IOH - Unit': 'IOH - Base Unit Count', 'IOH - On Hand Value':'SW Value OH', 'IOH - Qty On Hand':'SW Qty OH', 'IOH - Current Status':'SW Current Status Count',
                'IOH - No Repl Flag': 'SW No Repl Flag Count', 'IOH - LTM Demand Qty':'SW LTM Demand Qty', 'IOH - Reorder Point':'SW Avg Reorder Point', 
                'Base Reorder Point':'Base Reorder Point (Total)'}

# call rename () method
df_latestioh.rename(columns= rename_dict, inplace=True)
```

```python

# IOH Base Count Ck
df_latestioh['IOH - Base Count Ck'] = df_latestioh.apply(lambda x: x['Part in CS01/02']+x['Part in CS04'], axis = 1)

df_latestioh['SW Qty OH (Rail+Bus+CS)'] = df_latestioh.apply(lambda x: x['Rail OH Quantity']+x['Bus OH Quantity']+x['CS OH Quantity'], axis = 1)

df_latestioh['SW Value OH (Rail+Bus+CS)'] = df_latestioh.apply(lambda x: x['Rail OH Value']+x['Bus OH Value']+x['CS OH Value'], axis = 1)

df_latestioh['Global Days on Hand'] = df_latestioh.apply(lambda x: round(x['SW Qty OH (Rail+Bus+CS)'] / x['CS LTM Demand Quantity'] / 360, 0) if x['CS LTM Demand Quantity']>0 else None, axis = 1)

df_latestioh['DOH Duration'] = df_latestioh.apply(lambda x: duration_apply(x['Global Days on Hand'], x['SW Value OH (Rail+Bus+CS)']), axis = 1)

df_latestioh['DOH Status'] = df_latestioh.apply(lambda x: 'Inactive' if (x['SW LTM Demand Qty']==0 and x['SW LTM Demand Qty'] is None) else 'Active', axis = 1)

df_latestioh['Part in a Central Store'] = df_latestioh.apply(lambda x: x['Part in CS04']>0 or x['Part in CS01/02']>0, axis = 1)

df_latestioh['Central Store Category'] = df_latestioh.apply(lambda x: cs_category_apply(x['Part in CS01/02'], x['Part in CS04']), axis = 1)

df_latestioh['Descr Contains Do Not Reorder'] = df_latestioh['IOH - Item Descr'].apply(lambda x: 'Do Not Reorder' in x or 'Do Not Replen' in x)
```

```python
#left join
df_latestioh_merged = df_latestioh.merge(df_cs04reorder, left_on='IOH - Item ID', right_on='item_id', how='left')
```

```python
def do_not_repl_rule_apply(cs_category, contains_dnr, sw_no_repl_count, base_no_repl_count, base_unit_count, sw_ltm_qty, sw_qty_oh, 
                                avg_base_reorder, other_cs_count, other_cs_dnr_count, 
                                cs04_no_repl, cs04_cur, cs04_reorder_pt, cs04_reorder_qty, cs04_max_qty):
    if cs_category != 'In CS04':
        if contains_dnr:
            return 'Descr DNR'
        elif cs_category == 'Base Only':
            if sw_no_repl_count == base_no_repl_count:
                return 'No Repl Flag Set'
            elif sw_ltm_qty == 0 and sw_qty_oh == 0 and avg_base_reorder == 0:
                return 'SW OH/LTM/Reorder Qty = 0'
        else:
            if base_unit_count == 0 and (other_cs_count == other_cs_dnr_count):
                return 'No Repl Flag Set'
            elif other_cs_dnr_count == 0:
                return 'Replenish'
            elif other_cs_dnr_count == other_cs_count:
                return 'Other DNR Rule'
            else:
                return 'Investigate'
    else:
        if cs04_no_repl == 'Y':
            return 'No Repl Flag Set'
        elif cs04_cur == 'Discnt':
            return 'Current Status = Discnt'
        elif cs04_reorder_pt == 0 and cs04_reorder_qty == 0 and cs04_max_qty == 0:
            return 'Min/Max, Reorder Qty = 0'
        else:
            return 'Replenish'

df_latestioh_merged['CS04 Do Not Replenish Rule'] = df_latestioh_merged.apply(lambda x: do_not_repl_rule_apply(x['Central Store Category'], x['Descr Contains Do Not Reorder'], x['SW No Repl Flag Count'], x['Base Do Not Replenish Count'], x['Base Unit Count'],
                                                                                                    x['SW LTM Demand Qty'], x['SW Qty OH'], x['Avg Base Reorder Point'], x['Other CS Count'], 
                                                                                                    x['Other CS Do Not Replenish Count'], x['CS04 No Repl Flag'], x['CS04 Current Status'],
                                                                                                    x['CS04 Reorder Point'], x['CS04 Reorder Qty'], x['CS04 Max Qty'] ), axis = 1)
```

```python
def transport_class_apply(item_group, item_group_id):
    if item_group_id is None:
        return 'No Group Code'
    elif 'NABI' in item_group:
        return 'Bus - NABI'
    elif 'NEOPL' in item_group:
        return 'Bus - NeoPlan'
    elif any(s in item_group for s in ['New Flyer', 'NF 60', 'NewFly', 'NF', 'SR2206']):
        return 'Bus - New Flyer'  
    elif any(s in item_group for s in ['RTS', 'GMC']):
        return 'Bus - Nova RTS'
    elif 'WESTERN FLY' in item_group:
        return 'Bus - Western Flyer'
    elif 'GENERIC' in item_group:
        return 'General Material'
    elif 'TOOLS' in item_group:
        return 'General Material - Tools'
    elif 'PAINT' in item_group:
        return 'General Material - Paint'
    elif 'UPHOLST' in item_group:
        return 'General Material - Upholstery'
    elif 'PPE' in item_group:
        return 'Safety (PPE)'
    elif any(s in item_group for s in ['WORK CAR', 'CAMBRIDGE']):
        return 'Working Rail'
    elif any(s in item_group for s in ['BLUE LINE', 'BL']):
        return 'Rail - Blue Line'
    elif any(s in item_group for s in ['GREEN LINE', 'PCC']):
        return 'Rail - Green Line'
    elif 'CRRC' in item_group:
        return 'Rail - CRRC Orange/Red'
    elif 'ORANGE' in item_group:
        return 'Rail - Orange Line'
    elif 'RED LINE' in item_group:
        return 'Rail - Red Line'
    else:
        return item_group

def transport_segment_apply(trans_class):
    if 'Bus' in trans_class:
        return 'Bus'
    elif 'General' in trans_class:
        return 'General Material'
    elif 'Green Line' in trans_class:
        return 'Light Rail(Green Line)'
    elif 'Rail' in trans_class:
        if trans_class == 'Working Rail':
            return 'Working Rail'
        else:
            return 'Heavy Rail (Red/Orange/Blue Line)'
    elif 'Safety' in trans_class:
        return 'Safety Material'
    else:
        return None

def transport_category_apply(trans_class):
    if 'Bus' in trans_class:
        return 'Buses'
    elif 'General' in trans_class:
        return 'General Material'
    elif 'Rail' in trans_class:
        return 'Rail'
    elif 'Safety' in trans_class:
        return 'Safety Material'
    else:
        return None
```

```python
df_doh['DOH/Lead Time Ratio'] = df_doh.apply(lambda x: round(x['Global Days on Hand'] / x['Lead Time (Days)'], 1) if x['Lead Time (Days)']!= 0 else None, axis = 1)
df_doh['Transport Class'] = df_doh.apply(lambda x: transport_class_apply(x['IOH - FMIS Item Group ID'], x['IOH - FMIS Item Group']),axis = 1)
df_doh['Transport Segment'] = df_doh['Transport Class'].apply(transport_segment_apply)
df_doh['Transport Category'] = df_doh['Transport Class'].apply(transport_category_apply)
```
