```python
import pandas as pd
import numpy as np
```

```python
# read required excel sheets into dfs forcing string type
po_df = pd.read_excel (r'PL_PURCH_ORDER_DATA_PT1.xlsx', dtype=str) 

# change types of columns
po_df = po_df.astype({'PO - PO Hdr Match Process Inst': str, 
                      'PO - Query Date': 'datetime64[ns]', 
                      'PO - PO Last Update': 'datetime64[ns]', 
                      'PO - Contract Expire Date': 'datetime64[ns]',
                      'PO - FMIS DBE Goal': float, 
                      'PO - FMIS DBE Committed': float})
```

```python
# drop columns
po_df = po_df.drop(columns=['PO - Distrib Line Process Flag', 'PO - In Process Flag', 'PO - PO Post Status', 'PO - Dispatch Action', 'PO - Prepaid PO Flag',
       'PO - PO Prepaid Amt', 'PO - Match Flg'])
```

```python

def po_contract(contract, contract_id):
    if contract is None and contract_id is not None:
        if contract_id=="<NA>" or contract_id=="<NONE>":
            return None
        else:
            return contract_id
    elif contract=="<NONE>" or contract=="<NA>":
        return None
    else:
        if len(str(contract)) > len(str(contract_id)):
            return contract
        else:
            if contract_id=="<NA>" or contract_id=="<NONE>":
                return None
            else:
                return contract_id
                
# Calculated fields
po_df["CALC - PO Requesting Dept"] = pd.Series([row["PO - Dept ID"] if pd.isnull(row["PO - FMIS Requesting Dept ID"]) else row["PO - FMIS Requesting Dept ID"] for idx, row in po_df.iterrows()])
po_df["PO - Budget Status Descr"] = np.select(condlist=[po_df["PO - Budget Status"]=='E', po_df["PO - Budget Status"]=='N', po_df["PO - Budget Status"]=='V'],
                                                choicelist=['Budget Error', 'Not Checked', 'Valid'], default="")
po_df["CALC - PO Contract"] = po_df.apply(lambda x: po_contract(x["PO - FMIS Contract"], x["PO - FMIS Contract ID"]), axis = 1)

rename_dict = {'PO - Bid Categ Id': 'PO - FMIS Bid Categ ID', 'PO - FMIS Dept Short Descr': 'PO - FMIS Dept Short Descr', 'PO - FMIS Dept Descr': 'PO - Dept Descr'}

# call rename () method
po_df.rename(columns= rename_dict, inplace=True)
```
